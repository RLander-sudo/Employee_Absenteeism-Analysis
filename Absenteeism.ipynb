```python
import pandas as pd
# loading the dataset
df = pd.read_excel("Absenteeism_at_work.xls")

# Displaying loaded dataset, first few rows only
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns.values
```

```python
# renaming columns for better and quicker readability
df.columns = ['ID', 'Reason_for_absence', 'Month', 'Day_of_week',
       'Seasons', 'Transportation_expense',
       'Distance_to_Work', 'Service_time', 'Age',
       'Workload_Average_day ', 'Hit_target', 'Disciplinary_failure',
       'Education', 'Son', 'Social_drinker', 'Social_smoker', 'Pet',
       'Weight', 'Height', 'BMI', 'Absenteeism_time']

#checking data types
df.info()
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# What is the most common reason for being absent
plt.figure(figsize = (10, 5))
sns.countplot (x = df['Reason_for_absence'], order = df['Reason_for_absence'].value_counts().index, hue = df['Reason_for_absence'], palette = "viridis")
plt.title("Top Reasons for Employee Absenteeism")
plt.xlabel("Reason code")
plt.ylabel("Count")
plt.show()
```

```python
# How does the age affect absenteeism of an employee

plt.figure(figsize = (8,5))
sns.boxplot(x = df['Age'], y = df['Absenteeism_time'], hue = df['Age'], palette = "coolwarm")
plt.title("Absenteeism Time vs. Age")
plt.show()
```

```python
# which month have the most absenteeism

plt.figure(figsize = (10, 5))
sns.barplot(x = df['Month'], y= df['Absenteeism_time'], errorbar = None, hue = df['Month'], palette = 'magma')
plt.title("Average Absenteeism Time per Month")
plt.xlabel("Month")
plt.ylabel("Absenteeism Hours")
plt.show()
```
